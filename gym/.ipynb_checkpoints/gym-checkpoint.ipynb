## 3. Building a model for predicting customer churn

Target attribute - the fact of customer churn in the next month

```python
X = data.drop('churn', axis=1)
y = data['churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0,  stratify = y )
```

```python
log_reg_model = LogisticRegression(random_state=0, max_iter=5000, solver='lbfgs')
log_reg_model.fit(x_train,y_train)
log_reg_predictions = log_reg_model.predict(x_test)

dec_tree_model = DecisionTreeClassifier(random_state=0)
dec_tree_model.fit(x_train,y_train)
dec_tree_predictions = dec_tree_model.predict(x_test)
```

```python
print('LogisticRegression metrics:')
print('Accuracy - {:.2f}'.format(accuracy_score(y_test, log_reg_predictions)))
print('Precision - {:.2f}'.format(precision_score(y_test, log_reg_predictions)))
print('Recall - {:.2f}'.format(recall_score(y_test, log_reg_predictions)))

print('DecisionTreeClassifier metrics:')
print('Accuracy - {:.2f}'.format(accuracy_score(y_test, dec_tree_predictions)))
print('Precision - {:.2f}'.format(precision_score(y_test, dec_tree_predictions)))
print('Recall - {:.2f}'.format(recall_score(y_test, dec_tree_predictions)))
```

Logistic regression is more accurate in all parameters

## 4. Clustering of users

```python
scaler = StandardScaler()
scaler.fit(X)
X_st = scaler.transform(X)
```

```python
linked = linkage(X_st, method = 'ward') 
```

```python
plt.figure()  
dendrogram(linked, orientation = 'top')
plt.title('Clustering of fitness club clients')
plt.show()
```

Using the visualization algorithm we get 4 clusters (by color) and 5 by condition - probably the right cluster breaks into 2

```python
scaler = StandardScaler() 
data_sc = scaler.fit_transform(data) 
linked = linkage(data_sc, method = 'ward')
plt.figure()  
dendrogram(linked, orientation='top')
plt.title('Clustering of fitness club clients with regard to churn')
plt.show()
```

```python
km = KMeans(n_clusters = 5, random_state = 0)
k_means_clusters = km.fit_predict(X_st)
data['clusters'] = k_means_clusters
```

### 4.1 Examination of the average value of the cluster features

```python
data.groupby('clusters').mean().sort_values(by='churn')
```

* 1 cluster (churn 12%) - close to the location, a partner of the club, brought friends, half of the training consists of group
* 2 cluster (churn 20%) - close to location, one quarter are partners of the club, they didn't bring their friends, all workouts are group 
* 3 cluster (churn 27%) - a little farther from location, half are club partners, a third brought friends, half of the workouts consist of group workouts
* 4 cluster (38% churn) - close to location, a quarter are club partners, did not bring friends, do not go to group workouts
* 0 cluster (40% churn) - far from location, half are club partners, do not bring friends, a quarter of workouts are group

### 4.2 Distribution of features for clusters

```python
for value in [0, 1, 2, 3, 4]:
    data[data['clusters']==value].hist(figsize=(17,10))
    plt.suptitle('\nDistribution of features for the {} cluster'.format(value), fontsize = 15)
    plt.show()
```

## 5. Conclusions and basic recommendations

- Age and gender do not affect the length of the contract.

At the same time, if the remaining number of months decreases, the probability of renewal decreases, it is worth:
* Conduct additional work with clients who are running out of subscriptions. 
That is, the program "subscription" with a monthly extension in this case does not work, you should try to sell immediately for 6 months (discounts, bonuses, additional preferences).

In this case, apparently, group classes are a hook for customers. Accordingly, the suggestion is:
* Do a customer survey of the strengths and weaknesses of group classes and work through them, making them even better;
* Find out why the gym itself without group classes doesn't appeal to customers and work out the weaknesses (few specific machines, small showers, no cooler, etc.)
